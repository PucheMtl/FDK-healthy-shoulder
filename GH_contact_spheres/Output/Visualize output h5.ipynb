```python
import h5py
import numpy as np
```

```python
file_path = 'output.anydata.h5'

def print_all_hdf5_items(item, prefix=''):
    if isinstance(item, h5py.Dataset):
        print(f"{prefix}[Dataset]: {item.name}")
    elif isinstance(item, h5py.Group):
        print(f"{prefix}[Group]: {item.name}")
        for key in item.keys():
            print_all_hdf5_items(item[key], prefix + "  ")
            
```

```python
# Open the HDF5 file
h5_file = h5py.File(file_path, 'r')
```

```python
# Traverse the file's structure and print all datasets and groups
print_all_hdf5_items(h5_file)

# Close the HDF5 file
h5_file.close()
```
